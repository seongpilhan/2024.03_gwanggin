```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split # dataset split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error, r2_score # 평가도구
from sklearn.metrics import confusion_matrix, accuracy_score # model 평가
import pandas as pd

df = pd.read_csv(r'/서울시 상권분석서비스(폐업여부).csv', encoding='CP949')
df.info()
```

```python
cols = list(df.columns)
cols.pop(0)
cols.pop(2)
x_cols = cols

# 2번 열에 대해 레이블링을 수행하여 y_col 생성
y_col = df.iloc[:, 1].apply(lambda x: 1 if x >= 70 else 0)

# X와 y 설정
X = df[x_cols]
y = y_col
print(y)
```

```python
# 단계3. train/test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size = 0.3, random_state=34)
```

```python
# 단계4. model 생성
model = LogisticRegression(solver = 'lbfgs', max_iter=100, random_state = 1).fit(X=X_train , y=y_train)
```

```python
# 단계5. model 평가
y_pred = model.predict(X_test) # 예측치
y_true = y_test # 관측치
```

```python
# 1) 혼동행렬
con_max = confusion_matrix(y_true, y_pred)
print(con_max)
```

```python
# 2) 분류정확도
acc = accuracy_score(y_true, y_pred)
print(acc)
```
